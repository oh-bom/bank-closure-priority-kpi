시간대 09~16시로 필터링


```python
df=original
df=df[(df["시간대구분"]>=9)&(df["시간대구분"]<=16)]
df=df.reset_index()
df
```

```python
#기준일 하루로 추출
df=df[df["기준일ID"]==20240416]
df
```

필요 없는 컬럼들 삭제

```python
df=df.drop("index",axis=1)
df
```

```python
col_size=len(df.columns)
col_size
```

```python
temp_df=df
len(temp_df.columns)
```

```python
age_band=10
#10대가 두개로 나뉜건 20~60대, 70대는 하나로
gap=14
#남자: 5 , 여자: 19부터 시작
for i in range(5,18,2):
  #70대인 경우
  if i==17:
    temp_df[str(age_band)+"대"]=temp_df.iloc[:,i]+temp_df.iloc[:,i+gap]

  else:
    temp_df[str(age_band)+"대"]=temp_df.iloc[:,i]+temp_df.iloc[:,i+1]+temp_df.iloc[:,i+gap]+temp_df.iloc[:,i+1+gap]

  age_band+=10



temp_df
```

```python
#필요 없는 앞쪽 인덱스 삭제
re=temp_df.drop(temp_df.columns[4:col_size],axis=1)
re
```
