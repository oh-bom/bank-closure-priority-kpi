```python
by_asset.drop(by_asset.columns[0],axis=1,inplace=True)
by_asset
```

```python
asset_rank=by_asset.loc[:,["구간","등급"]]
asset_rank
```

```python
def get_rank(x):
  x*=10000
  for index,row in asset_rank.iterrows():
    if x>500000000:
      label="a"
      break
    else:
      label=row["등급"] #우선 등급 부여
      if x<=row["구간"]:break

  return label
```

```python
v11["노년층연평균소득합본"]=v11["노년층연평균소득"]+v11["노년층연평균소득추정"]

v11["노년층자산규모등급"]=v11.loc[:,"노년층연평균소득합본"].apply(lambda x:get_rank(x))
```

### 청장년층 시간 불편도 계산

교통비, 최저 시급은 상수
시간:자택에서 지점 간 택시비(대중 교통비) * 방문 횟수 + 2024년 최저시급(9860원) * 대기 시간

```python
FEE=1500
Hourly_Wage=9860
```

```python
v12=v11
by_age
```

```python
young_visit=by_age.loc[by_age["분류"]=="청장년층"]["방문횟수"].iat[0]
young_waiting=by_age.loc[by_age["분류"]=="청장년층"]["대기시간"].iloc[0]

young_timeloss_cost=FEE*young_visit+Hourly_Wage*young_waiting/60
young_timeloss_cost
```

```python
v12["청장년층_시간불편도"]=v12["2050대_인구수"]*young_timeloss_cost
```

시간:자택에서 지점 간 택시비(대중 교통비) * 방문 횟수 + 2024년 최저시급(9860원) * 대기 시간

```python
v_t=by_asset.loc[:,["등급","방문횟수","대기시간"]]
```

```python
v_t[v_t["등급"]=="b"]["대기시간"].iloc[0]
```

```python
v13["6070인구"]=v13["60대"]+v13["70대"]
v13["6070인구"]
```

```python
def calc_old_loss(x):
  rank=x[0]
  old_cnt=x[1]

  visit=v_t[v_t["등급"]==rank]["방문횟수"].iloc[0]
  wait=v_t[v_t["등급"]==rank]["대기시간"].iloc[0]
  cost=FEE*visit+Hourly_Wage*wait

  total_time_loss_cost=cost*old_cnt

  return total_time_loss_cost
```

```python
v13["노인층_시간불편도"]=v13[["노년층자산규모등급","6070인구"]].apply(calc_old_loss,axis=1)
```
