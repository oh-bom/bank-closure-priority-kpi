```python
seoul_code=code_df.drop('생성일자',axis=1)
seoul_code=seoul_code.drop('말소일자',axis=1)
# seoul_code=seoul_code.loc[:,"행정동코드"].apply(lambda x:x/100)
seoul_code["행정동코드"]=seoul_code["행정동코드"].apply(lambda x:int(x/100))
seoul_code
```

```python
path2="/content/drive/MyDrive/asac/project/eda/서울시 상권분석서비스(소득소비-행정동).csv"
income_df = pd.read_csv(path2, encoding='cp949')


income_df=income_df.iloc[:,:5] # 필요 없는 열들 제거
income_df
```

```python
type(income_df["기준년분기코드"][0])
```

서울행정동 코드+ 행정동별 월평균소득 데이터 머지


```python
merged_df=pd.merge(seoul_code,income_df,
         how="right",
         left_on="행정동코드",
         right_on="행정동코드"
)
merged_df
```

```python
# null 확인
null_data=merged_df["월평균소득"].isna()
null=df[null_data]
null
```


각 행정동코드당 몇개의 데이터 있는지 확인

```python
#행정동 코드당 데이터 개수 확인
income_avg_cnt=income_df.groupby(by=["행정동코드","행정동명"])["월평균소득"].count()
income_avg_cnt=income_avg_cnt.reset_index()

#19개가 아닌경우 있는지 확인
filter=income_avg_cnt[income_avg_cnt["월평균소득"]!=19]
display(income_avg_cnt)

print(filter)
```

```python
#행정동 코드당 데이터 개수 확인
income_avg_cnt=merged_df.groupby(by=["행정동코드","시군구명","행정동명"])["월평균소득"].count()
income_avg_cnt=income_avg_cnt.reset_index()

#19개가 아닌경우 있는지 확인
filter=income_avg_cnt[income_avg_cnt["월평균소득"]!=19]
display(income_avg_cnt)

print(filter)
```

```python
#행정동당 월평균 소득
income_avg2=income_df.groupby(by=["행정동코드","행정동명"])["월평균소득"].mean().reset_index()
income_avg2
```

```python
#행정동당 월평균 소득
income_avg=merged_df.groupby(by=["행정동코드","시군구명","행정동명"])["월평균소득"].mean().reset_index()
income_avg
```

```python
temp=pd.pivot_table(merged_df,
               index=["시군구명","행정동명","행정동코드"],
               values=["월평균소득"],
               )
temp
```
